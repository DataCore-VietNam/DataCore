## 1. Initialize the client

```python
from datacore import Datacore

# Demo mode
client_demo = Datacore()

# Paid mode

x_api_key = "your_api_key" #Fill in your Datacore API key
client_paid = Datacore(api_key=x_api_key)
```

## 2. Preview a dataset (Demo mode)

```python
# All columns
df_demo = client_demo.preview("dataset_historical_price")
df_demo
```

```python
print(df_demo.head(5))
```

```python
# Filter specific columns
df_demo = client_demo.preview("dataset_historical_price", columns=["symbol", "date", "close_price"])
df_demo
```

## 3. Fetch data (Paid mode)

```python
# All columns — returns {"data": DataFrame, "info": str}

result = client_paid.get_data("dataset_historical_price")
print(result["data"])
print(result["info"])
```

```python
# Filter specific columns
result = client_paid.get_data("dataset_historical_price", columns=["symbol", "date", "close_price"])
print(result["data"])
print(result["info"])
```

## 4. Download data to file

```python
# Download all pages
download_result = client_paid.download_data(
    dataset_code="dataset_historical_price",
    output_path="dataset_historical_price_all.csv",  # hoặc .json
    file_format="csv",   # "csv" hoặc "json"
    start_page=1,
    end_page=None,       # để None là tải tới trang cuối
    show_progress=True,
)
print(download_result)
```

```python
# Download only first 3 pages
download_result = client_paid.download_data(
    dataset_code="dataset_historical_price",
    output_path="dataset_historical_price_page1_3.json",
    file_format="json",  
    start_page=1,
    end_page=3,       
    show_progress=True,
)
print(download_result)
```

```python
# select columns to download
download_result = client_paid.download_data(
    dataset_code="dataset_historical_price",
    columns=["symbol", "date", "close_price"],
    output_path="dataset_historical_price_selected_cols.csv",
    file_format="csv",
    start_page=1,
    end_page=3,
    show_progress=True,
)
print(download_result)
```

```python
# Download page 8 only
download_result = client_paid.download_data(
    dataset_code="dataset_historical_price",
    columns=["symbol", "date", "close_price"],
    output_path="dataset_historical_price_page8.csv",
    file_format="csv",
    start_page=8,
    end_page=8,
    show_progress=True,
)
print(download_result)
```
